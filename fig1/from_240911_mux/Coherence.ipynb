#### Kerrs and Cross Kerr

```python
kerrs = np.array([coupler_anharmonicity * (freq_gs[1][idx]*1e-3 / (coupler_freq - freq_gs[0][idx]))**4 for idx in range(len(freq_gs[0]))])
plt.plot(freq_gs[0], kerrs*1e+6, 'o-')
plt.yscale('log')
```

```python
cross_kerr_matrix = np.zeros((len(freq_gs[0]), len(freq_gs[0])))
for mode_idx in range(len(freq_gs[0])):
    for mode_jdx in range(len(freq_gs[0])):
        cross_kerr_matrix[mode_idx][mode_jdx] = 2* coupler_anharmonicity * (freq_gs[1][mode_idx]*1e-3 / (coupler_freq - freq_gs[0][mode_idx]))**2 * (freq_gs[1][mode_jdx]*1e-3 / (coupler_freq - freq_gs[0][mode_jdx]))**2
        if mode_idx != mode_jdx:
            cross_kerr_matrix[mode_idx][mode_jdx] *= 0.5
cross_kerr_matrix
```

# Drawing pulse shape

```python

# Define a function to create a Gaussian pulse
def gaussian_pulse(t, amplitude, center, width):
    cutoff_start = center - 2 * width
    cutoff_end = center + 2 * width
    pulse = np.zeros_like(t)
    mask = (t >= cutoff_start) & (t <= cutoff_end)
    pulse[mask] = amplitude * np.exp(-(t[mask] - center)**2 / (2 * width**2))
    return pulse

# Define a function to create a Gaussian flat-top pulse with 2-sigma ramps
def gaussian_flat_top_pulse(t, amplitude, start, flat_length, ramp_sigma):
    ramp_length = 2 * ramp_sigma  # 2 sigma cutoff for the ramp
    end = start + flat_length + 2 * ramp_length  # Total length of the pulse
    pulse = np.zeros_like(t)

    # Left ramp (left half of Gaussian)
    left_ramp = (t >= start) & (t < start + ramp_length)
    pulse[left_ramp] = amplitude * np.exp(-((t[left_ramp] - start - ramp_length)**2) / (2 * ramp_sigma**2))

    # Flat top
    flat_top = (t >= start + ramp_length) & (t < start + ramp_length + flat_length)
    pulse[flat_top] = amplitude

    # Right ramp (right half of Gaussian)
    right_ramp = (t >= start + ramp_length + flat_length) & (t < end)
    pulse[right_ramp] = amplitude * np.exp(-((t[right_ramp] - (start + ramp_length + flat_length))**2) / (2 * ramp_sigma**2))

    return pulse

# Define the parameters
t = np.linspace(0, 160, 1000)  # Time axis
channels = ['Qubit', 'Coupler RF']  # List of channels
channel_spacing = 1.5  # Vertical spacing between channels
pulses = {
    'Qubit': [
        
        {'type': 'gaussian', 'center': 8, 'width': 3, 'amplitude': 0.3},
        {'type': 'gaussian', 'center': 21, 'width': 3, 'amplitude': 0.5},
        {'type': 'square', 'start': 28, 'end': 60, 'amplitude': 0.6},
        {'type': 'gaussian', 'center': 92, 'width': 3, 'amplitude': 0.5},
        {'type': 'gaussian', 'center': 124, 'width': 3, 'amplitude': 0.5},
        # {'type': 'gaussian', 'center': 92+60, 'width': 3, 'amplitude': 0.5},
        # {'type': 'gaussian', 'center': 124+60, 'width': 3, 'amplitude': 0.5}

    ],
    'Coupler RF': [
        {'type': 'square', 'start': 61, 'end': 80, 'amplitude': 0.6},
        # {'type': 'square', 'start': 130, 'end': 140, 'amplitude': 0.6},
        
    ]
}

# Assign a unique color for each channel
channel_colors = {
    'Qubit': 'red',
    'Coupler RF': 'Purple'
}

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot pulses on the same plot with vertical offsets for different channels
for i, channel in enumerate(channels):
    offset = i * channel_spacing  # Vertical offset for each channel
    color = channel_colors[channel]  # Get the color for the current channel
    
    # Plot pulses for each channel
    for pulse in pulses[channel]:
        if pulse['type'] == 'square':
            start, end, amp = pulse['start'], pulse['end'], pulse['amplitude']
            pulse_shape = np.where((t >= start) & (t <= end), amp, 0) + offset
            ax.plot(t, pulse_shape, label=f'{channel} - Square Pulse', color=color)
            ax.fill_between(t, offset, pulse_shape, where=(t >= start) & (t <= end), color=color, alpha=0.3)
        elif pulse['type'] == 'gaussian':
            center, width, amp = pulse['center'], pulse['width'], pulse['amplitude']
            pulse_shape = gaussian_pulse(t, amp, center, width) + offset
            ax.plot(t, pulse_shape, label=f'{channel} - Gaussian Pulse', color=color)
            ax.fill_between(t, offset, pulse_shape, color=color, alpha=0.3)
        elif pulse['type'] == 'gaussian_flat_top':
            start, flat_length, ramp_sigma, amp = pulse['start'], pulse['flat_length'], pulse['ramp_sigma'], pulse['amplitude']
            pulse_shape = gaussian_flat_top_pulse(t, amp, start, flat_length, ramp_sigma) + offset
            ax.plot(t, pulse_shape, label=f'{channel} - Gaussian Flat-Top Pulse', color=color)
            ax.fill_between(t, offset, pulse_shape, color=color, alpha=0.3)

# Labeling and styling
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude (Shifted)')
ax.set_title('Pulse Control Diagram with Gaussian Flat-Top Pulse')
# ax.legend()
# ax.grid(True)

# Set y limits to make sure all pulses are visible
ax.set_ylim(-0.5, len(channels) * channel_spacing)

# Hide the x and y axes
ax.axis('off')

plt.show()
```
